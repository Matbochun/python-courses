```python
def multiply_sum(items):
    total = 1
    for x in items:
        total *= x    #multiplies total by item x, ten again and again in next iteration
    return total
        
```

```python
multiply_sum([1,2,4])
```

```python
multiply_sum([6,15,28])
```
