```python
sp500.loc['2013-1-1': '2013-12-31', 'close'].plot(style='k.-', title='S&P 500') # k = color(here: black)/. stands for indication/- stand for line style
plt.ylabel('Closing price')
plt.show()
```

```python
sp500['close'].plot(title='S&P 500', kind='area')
plt.ylabel('Closing price')
plt.show()
```
